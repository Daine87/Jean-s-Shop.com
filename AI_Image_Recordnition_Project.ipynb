```python
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, decode_predictions, preprocess_input
from PIL import Image

# Upload an image
uploaded = files.upload()

# Load the model
model = MobileNetV2(weights="imagenet")

# Define image preprocessing function
def preprocess_image(image):
    image = image.resize((224, 224))  # Resize correctly
    image = np.array(image) / 255.0  # Normalize between 0 and 1
    image = np.expand_dims(image, axis=0)  # Add batch dimension
    return image

# Load and display the image
for filename in uploaded.keys():
    img = Image.open(filename)
    plt.imshow(img)
    plt.axis("off")
    plt.show()

    # Use the function before making predictions
    img_array = preprocess_image(img)

    # Make a prediction
    predictions = model.predict(img_array)
    decoded_predictions = decode_predictions(predictions, top=1)[0]  # Get only the top 1 prediction

    label = decoded_predictions[0][1]  # Extract the label
    confidence = decoded_predictions[0][2] * 100  # Convert confidence to percentage

    print(f"Predicted Object: {label} ({confidence:.2f}%)")
```
